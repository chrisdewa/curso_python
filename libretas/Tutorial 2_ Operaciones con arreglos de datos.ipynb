## Acceso a variables (columnas)
Hay dos formas de acceder a las variables del arreglo de datos. Una es con punto y el nombre de la variable y la otra es con la notación de diccionario


### Punto y nombre (atributo)
En python, los atributos son propiedades de los objetos a los que podemos acceder con punto pero sin paréntes (no son funciones que se ejecutan).

`df.peso`

Esta notación solo funciona cuando la variable tiene un nombre válido en python. En este arreglo de datos todas las variable pueden ser accesadas así.

Esta notación tiene limitantes (ver tutorial 1).

````python
df.HAS.head()
````

````python
df.peso
````

### Notación de diccionario

En un diccionario definido como:
```python
diccionario = {'a': 1, 'b': 2}
```

podemos acceder a las "llaves" con corchetes:

```python
diccionario['a'] == 1 # True
```

En pandas, podemos acceder a las columnas de la misma manera:


````python
df['peso'].head()
````
